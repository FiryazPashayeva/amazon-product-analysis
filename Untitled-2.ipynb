```python
import numpy as np
import pandas as pd
from scipy import stats
from scipy.stats import shapiro
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\pasha\Downloads\archive\amazon.csv")
df.head()
```

```python

nan_counts = df.isna().sum()
print("Number of NaN values per column")
print(nan_counts)
print('Number of dublicated values:',df.duplicated().sum())
print(df.columns.tolist())
print(df.head())

```

```python
df['discounted_price'] = (
    df['discounted_price']
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
    .astype(float)
)

df['actual_price'] = (
    df['actual_price']
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
    .astype(float)
)
```

```python
df.info()
print(df.dtypes)
```

```python
df['discount_percentage'] = (
    df['discount_percentage']
    .str.replace('%', '', regex=False)
    .astype(int)
)
```

```python
print(df['rating'].unique())
```

```python
df[df['rating'] == '|']
df = df[df['rating'] != '|']
df['rating'] = pd.to_numeric(df['rating'])
df.info()
```

```python
df['rating_count'] = (
    df['rating_count']
    .str.replace(',', '', regex=False)
)

df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')
```

```python
print("Skewness:", df['rating_count'].skew())
print("Kurtosis:", df['rating_count'].kurt())

stat, p = shapiro(df['rating_count'].dropna().sample(500, random_state=42))

print("Statistic:", stat)
print("p-value:", p.round())
```

```python
df['rating_count'] = df['rating_count'].fillna(df['rating_count'].median())
```

```python
df['Main_Category'] = df['category'].str.split('|').str[0]

df['Savings'] = df['actual_price'] - df['discounted_price']
```
